```python
from ipywidgets import widgets, interact, interactive
#from IPython.html.widgets import *
import numpy as np
import matplotlib.pyplot as plt
#import bqplot.pyplot as plt

z = 2
F = 96485
kb = 1,38E-23
R = 8.314
T = 298
j0 = 2.0E-0
jl = 10

A = z*F/(R*T)
#alpha = 0.5 # np.arange(0, 1, 0.1)
overpotential = np.arange(-0.07, 0.07, 0.001)

```

```python
%%latex
\begin{align}
j = j_C + j_A\\
= j_o(exp((1-\alpha) * A * overpotential)\\
\end{align}
```

```python
def Butler(alpha, j=1.0E-6):
    jA = j*np.exp((1-alpha) * A * overpotential)
    jC = -j*np.exp(-alpha * A * overpotential)
    jo = (jA + jC)

   # plt.figure(1)
   # plt.subplot(3,1,1)
   # plt.plot(overpotential,jA)
   # plt.subplot(3, 1, 2)
   # plt.plot(overpotential, jC)
   # plt.subplot(3, 1, 3)
   # plt.plot(overpotential, jo)

    fig = plt.figure(1)
    ax = fig.add_subplot(111)
    ax.plot(overpotential, jA)
    ax.plot(overpotential, jC)
    ax.plot(overpotential, jo)
    ax.set_xlim(-0.1, 0.1)
    ax.set_ylim(-30, 30)
    plt.show()

interact(Butler, alpha=(0.0, 1.0, 0.01), j = (1.0, 4.0, 0.1))
```
